Suma de las potencias de 2, desde 0 hasta n.

```python
def f1(num):
  if num == 0:
    return 1
  return f1(num - 1) + 2**num
```

Suma de las potencias de cualquier n√∫mero real, desde 1 hasta n.

```python
def f2(n,q):
  if n == 0:
    return 0
  return f2(n-1,q) + q**n 
```

Funci√≥n eAprox que aproxima el valor del n√∫mero e.

```python
def factorial(num):
  if num == 0:
    return 1
  else:
    return num * factorial(num - 1) 

def eAprox(num):
  if num == 0:
    return 1
  else:
    return eAprox(num - 1) + 1 / factorial(num) 

e = eAprox(50)
print(f'e: {e}')
```

Implementar una funci√≥n sumaRacionales(n,m) que sume todos los n√∫meros racionales de la forma p/q con 1 ‚â§ p ‚â§ n y 1 ‚â§ q ‚â§ m.

```python
def sumatoria(n):
  if n == 0:
    return 0
  return sumatoria(n-1) + n 

def sumaRacionales(n,m):
  if m == 0:
    return 0
  return sumaRacionales(n,m-1) + (1 / m) * sumatoria(n) 
```

```python
def f2(n,q):
  if n == 0:
    return 0
  return f2(n-1,q) + q**n

def f(n,m):
  if n == 0:
    return 0
  return f(n-1,m) + f2(m,n)

def h(n):
  if n == 1:
    return 0
  return h(n-1) + f2(n-1,n)

#Definici√≥n de la funci√≥n g2.
def g2(n):
  return f(n,n) - h(n)
```

Implementar una funci√≥n que dado un n, sume todos los n√∫meros naturales menores o iguales que n que tengan todos los d√≠gitos iguales.

***Funci√≥n digIguales***
*Supongamos que n = d1d2...d(k-1)dk. Entonces, me pregunto...¬øes dk == d(k-1)?.
Si la respuesta es s√≠, aplico la funci√≥n al n√∫mero d1d2...d(k-1)(ie, me pregunto...¬øes d(k-1) == d(k-2)?) Por transitividad, si d(k-1) == d(k-2) entonces dk == d(k-2). Y as√≠ sucesivamente.
El corte de la recursi√≥n lo da la condici√≥n n < 10.*


```python
def digIguales(num):
  return num < 10 or ((num % 10) == ((num // 10) % 10) and digIguales(num // 10))

def sumaDigIguales(num):
  if num < 10: #Corte de control. Sino suma todos los n√∫meros del 1 al 9.
    return 0
  else:
    if digIguales(num):
      return sumaDigIguales(num - 1) + num
    return sumaDigIguales(num - 1) 
```
