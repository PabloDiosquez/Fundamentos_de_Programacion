#**Pila 🔋**

Pila.png

```python
class _Nodo:
  def __init__(self, dato, prox):
    self.dato = dato
    self.prox = prox
```

```python
class Pila:
  def __init__(self):
    self.tope = None

  def apilar(self, dato):
    nuevo = _Nodo(dato, self.tope)
    self.tope = nuevo

  def desapilar(self):
    '''
    Pre: La pila NO está vacía
    Desapila un elemento y lo devuelve.
    '''
    if self.tope is None:
      raise ValueError('La pila está vacía')
    dato = self.tope.dato
    self.tope = self.tope.prox
    return dato

  def ver_tope(self):
    '''
    Pre: La pila NO está vacía.
    Devuelve el elemento al tope.
    '''
    if self.tope is None:
      raise ValueError('La pila está vacía')
    return self.tope.dato

  def esta_vacia(self):
    return self.tope is None

p = Pila()
p.esta_vacia()
# p.ver_tope()
p.apilar(4)
p.esta_vacia()
p.ver_tope()
p.apilar(7)
p.apilar(2)
p.ver_tope()
p.desapilar()
p.ver_tope()
```

#**Cola 🐴**

cola.png



```python
class Cola:
  def __init__(self):
    self.frente = None
    self.ultimo = None
  '''
    Invariantes:
    - Si la cola está vacía 👉 ambos son None
    - Si la cola tiene un solo elemento 👉 ambos apuntan al mismo nodo
    - Si la cola NO está vacía 👉 ambos no son None
  '''

  def encolar(self, dato):
    nuevo = _Nodo(dato, None)
    if self.frente is None:
      # la cola está vacía
      self.frente = nuevo
      self.ultimo = nuevo
    else:
      self.ultimo.prox = nuevo
      self.ultimo = nuevo

  def desencolar(self):
    '''
    Pre: la cola NO está vacía.
    Desencola el elemento que está al frente y lo devuelve.
    '''
    if self.frente is None:
      raise ValueError('La cola está vacía.')     
    dato = self.frente.dato
    self.frente = self.frente.prox
    # La cola quedó vacía, luego de desencolar?
    if self.frente is None:
      self.ultimo = None
    return dato

  def ver_frente(self):
    '''
    Pre: la cola NO está vacía.
    Devuelve el elemento que está al frente.
    '''
    if self.frente is None:
      raise ValueError('La cola está vacía.')     
    return self.frente.dato
  
  def esta_vacia(self):
    return self.frente is None

c = Cola()
c.esta_vacia()
# c.ver_frente()
c.encolar(4)
c.encolar(8)
c.desencolar()
c.ver_frente()
```
