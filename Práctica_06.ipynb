Se desea realizar un programa para una universidad. El mismo deberÃ¡ permitir guardar la

```python
class Curso:
  def __init__(self,nombre="",clases=0,dia="",hora=0,duracion=0):
    self.nombre = nombre
    self.clases = clases
    self.dia = dia
    self.hora = hora
    self.duracion = duracion
  
  def dia_de_semana(self):
    return self.dia

  def cantidad_clases(self):
    return self.clases

  def intensidad_del_curso(self):
    return self.clases*self.duracion

  def __str__(self):
    return f"""Nombre: {self.nombre} - Cantidad de clases: {self.clases} - DÃ­a: {self.dia} - Hora: {self.hora} - DuraciÃ³n: {self.duracion}
    """

  def primera_letra_nombre(self):
    return self.nombre[0]

def cargar_cursos():
  cursos = []
  c = int(input('Ingrese la cantidad de cursos: '))
  for i in range(c):
    print(f'Datos del curso {i+1}:')
    nombre = input('Nombre: ')
    clases = int(input('Cantidad de clases: '))
    dia = input('DÃ­a: ')
    hora = input('Hora: ')
    duracion = int(input('DuraciÃ³n: '))
    if duracion > 360: continue
 
    cursos.append(Curso(nombre,clases,dia,hora,duracion))
  return cursos

def datos_del_curso(lista,nombre):
  for e in lista:
    if e.nombre.lower() == nombre:
      return str(e)

def cursos_que_se_dictan_los(cursos,dia):
  for curso in cursos:
    if curso.dia.lower() == dia:
      print(curso.nombre)

def curso_mayor_intensidad(cursos):
  cursoMax = cursos[0].nombre
  mayorInt = cursos[0].intensidad_del_curso()
  for curso in cursos:
    if curso.intensidad_del_curso() > mayorInt:
      cursoMax = curso.nombre
      mayorInt = curso.intensidad_del_curso()
  return f'El curso de mayor intensidad es {cursoMax}'

def curso_empieza_con(cursos,letra):
  for curso in cursos:
    if curso.nombre[0].lower() == letra:
      print(curso.nombre)

def main():
  print('Cargar cursos:\n')
  cursos = cargar_cursos()
  print('Cursos con mÃ¡s de 5 clases:')
  for curso in cursos:
    if curso.clases > 5:
      print(curso.nombre)

  dia = input('Ingrese un dÃ­a de la semana: ')
  cursos_que_se_dictan_los(cursos,dia.lower())

  print(curso_mayor_intensidad(cursos))
    
  nombre = input('Ingrese un nombre: ')
  print(datos_del_curso(cursos,nombre.lower()))

  letra = input('Ingrese una letra: ')
  curso_empieza_con(cursos,letra.lower())
    
main()
```

- imprimir los datos de los artÃ­culos cuyo stock es 0
- imprimir la informaciÃ³n completa de los artÃ­culos que no requieren frÃ­o.

```python
class Articulo:
  def __init__(self,codigo,nombre,descripcion,precio_venta,proveedor,refrigeracion):
    self.codigo = codigo
    self.nombre = nombre
    self.descripcion = descripcion
    self.precio_venta = precio_venta
    self.refrigeracion = refrigeracion

def carga_articulos():
  articulos = []
  while True:
    codigo = int(input('CÃ³digo: '))
    nombre = input('Nombre: ')
    descripcion = input('Breve descripciÃ³n: ')
    precio_venta = float(input('Precio de venta: '))
    proveedor = input('Proveedor: ')
    refrigeracion = input('Requiere refrigeracion? (N -> NO -- R --> SÃ­): ')
    if codigo == 999 and refrigeracion =='N': break 

    articulos.append(Articulo(codigo,nombre,descripcion,precio_venta,proveedor,refrigeracion))
  
  return articulos

def stock(articulos):
  stock = {}
  for articulo in articulos:
    if articulo not in stock:
      stock[articulo.codigo] = 1
    else:
      stock[articulo.codigo] += 1

  return stock

def requiere_refrigeracion(articulos):
  refrigeracion = []
  print('ArtÃ­culos que necesitan refrigeraciÃ³n: ')
  for articulo in articulos:
    if articulo.refrigeracion == 'R':
      print(f'CÃ³digo: {articulo.codigo} - Nombre: {articulo.nombre}')
      refrigeracion.append(articulo)

  return refrigeracion  

def cantidad_con_refrigeracion(articulos):
  stock = stock(articulos)
  refrigeracion = requiere_refrigeracion(articulos)
  for articulo in refrigeracion:
    print(f'Cantidad: {stock[articulo.codigo]}')
  
```
