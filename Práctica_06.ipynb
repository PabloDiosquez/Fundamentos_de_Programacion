```python
class Equipo:
  def __init__(self, nombre, puntaje, goles):
    self.nombre = nombre
    self.puntaje = puntaje
    self.goles = goles
  
def cargar_equipos():
  equipos = []
  while True:
    nombre = input('Nombre: ')
    if nombre == 'fin': break
    puntaje = int(input('Puntaje: '))
    goles = int(input('Goles: '))

    equipos.append(Equipo(nombre,puntaje,goles))
  
  return equipos

def campeon(equipos):
  campeon = equipos[0]
  puntaje_max = equipos[0].goles

  for equipo in equipos:
    if equipo.puntaje > puntaje_max:
      campeon = equipo
      puntaje_max = equipo.puntaje
  
  return campeon

def main():
  equipos = cargar_equipos()
  print(f'Goles a favor del primer equipo de la lista: {equipos[0].goles}')
  print(f'Goles a favor del Ãºltimo equipo de la lista: {equipos[len(equipos) - 1].goles}')
  print(f'\nCampeÃ³n: {campeon(equipos).nombre}')
```

Ejercicio 4:
Se desea realizar un programa para una universidad. El mismo deberÃ¡ permitir guardar la
informaciÃ³n de todos los cursos que se dictan en la misma. palabrasa cantidad de cursos
existentes es ingresado por el usuario del sistema. De cada curso se debe conocer:
Nombre, Cantidad de Clases, DÃ­a de Dictado, Hora de Comienzo, DuraciÃ³n en minutos.
Importante: si la duraciÃ³n del curso es mayor a 360 minutos, NO se debe guardar la
informaciÃ³n del mismo.
El programa debe permitir

a) Imprimir todos los cursos que tengan mÃ¡s de 5 clases

b) Imprimir los cursos que se dictan en un dÃ­a de la semana indicado por el usuario.

c) Imprimir el curso mÃ¡s intensivo (definido por la cantidad de clases y la duraciÃ³n de
cada clase)

d) Imprimir los datos en que se dicta un curso indicado por el usuario.

e) Imprima todos los cursos que comienzan con una letra indicada por el usuario.

El programa escrito deberÃ¡ mostrar un menÃº para que el usuario elija la opciÃ³n deseada.
Importante: el sistema termina cuando el usuario elige la opciÃ³n â€œsalirâ€.

```python
class Curso:
  def __init__(self,nombre="",clases=0,dia="",hora=0,duracion=0):
    self.nombre = nombre
    self.clases = clases
    self.dia = dia
    self.hora = hora
    self.duracion = duracion
  
  def dia_de_semana(self):
    return self.dia

  def cantidad_clases(self):
    return self.clases

  def intensidad_del_curso(self):
    return self.clases*self.duracion

  def __str__(self):
    return f"""Nombre: {self.nombre} - Cantidad de clases: {self.clases} - DÃ­a: {self.dia} - Hora: {self.hora} - DuraciÃ³n: {self.duracion}
    """

  def primera_letra_nombre(self):
    return self.nombre[0]

def cargar_cursos():
  cursos = []
  c = int(input('Ingrese la cantidad de cursos: '))
  for i in range(c):
    print(f'Datos del curso {i+1}:')
    nombre = input('Nombre: ')
    clases = int(input('Cantidad de clases: '))
    dia = input('DÃ­a: ')
    hora = input('Hora: ')
    duracion = int(input('DuraciÃ³n: '))
    if duracion > 360: continue
 
    cursos.append(Curso(nombre,clases,dia,hora,duracion))
  return cursos

def datos_del_curso(lista,nombre):
  for e in lista:
    if e.nombre.lower() == nombre:
      return str(e)

def cursos_que_se_dictan_los(cursos,dia):
  for curso in cursos:
    if curso.dia.lower() == dia:
      print(curso.nombre)

def curso_mayor_intensidad(cursos):
  cursoMax = cursos[0].nombre
  mayorInt = cursos[0].intensidad_del_curso()
  for curso in cursos:
    if curso.intensidad_del_curso() > mayorInt:
      cursoMax = curso.nombre
      mayorInt = curso.intensidad_del_curso()
  return f'El curso de mayor intensidad es {cursoMax}'

def curso_empieza_con(cursos,letra):
  for curso in cursos:
    if curso.nombre[0].lower() == letra:
      print(curso.nombre)

def main():
  print('Cargar cursos:\n')
  cursos = cargar_cursos()
  print('Cursos con mÃ¡s de 5 clases:')
  for curso in cursos:
    if curso.clases > 5:
      print(curso.nombre)

  dia = input('Ingrese un dÃ­a de la semana: ')
  cursos_que_se_dictan_los(cursos,dia.lower())

  print(curso_mayor_intensidad(cursos))
    
  nombre = input('Ingrese un nombre: ')
  print(datos_del_curso(cursos,nombre.lower()))

  letra = input('Ingrese una letra: ')
  curso_empieza_con(cursos,letra.lower())
    
main()
```

```python
texto = """Ni bien amainÃ³
la tormenta, olvidÃ³
las promesas hechas
otra vez!
Me matan, LimÃ³n!
Hijueputas, LimÃ³n!
Por los techos viene el bloque
otra vez!
Oh, otra vez!
Oh, otra vez!
Pasan las dos... DelatÃ¡ndolo,
apenas las dos... delatÃ¡ndolo a las dos
de la tarde y zumba feo algÃºn moscÃ³n.
Voy sangrando, LimÃ³n! Padrecito de
Me la dieron, LimÃ³n! nuestros pobres que
sangran por las tejas como vos, LimÃ³n.
Oh, otra vez!
Oh, otra vez!
Nunca pude llorar,
ni soÃ±ar al dormir,
pero sabe que a las dos es tarde ya.
Todo el bloque, LimÃ³n, Nuestra gente, no,
baila y canta, mi Dios! no quiere escuchar
cumbias y merengues crueles otra vez!
Oh, otra vez!
Oh, otra vez!
Oh, otra vez!
Oh, otra vez!"""

def contador_palabras(texto):
  c = 0
  for carac in texto:
    if carac != ' ':
      c += 1
  return c

def contador_oraciones(texto):
  return len(texto.split('.\n'))

def contador_palabras_que_comienzan_con(texto,letra):
  palabras = texto.lower().split(' ')
  c = 0
  for p in palabras:
    if p and p[0] == letra: # when you split there will be an empty string at the end, and you can't access the first char char of this,
                            #  so just add a condition in the if:
                            # if word because 0-size word are False, and other True.
      c += 1
  return c 

def contador_de_palabra(texto,palabra):
  return texto.count(palabra)

def contador_palabras_comienzan_mayuscula(texto):
  palabras = texto.split(' ')
  c = 0
  for p in palabras:
    if p and p[0].isupper():
      c += 1
  return c

def contador_de_carac_numericos(texto):
  palabras = texto.split(' ')
  c = 0
  for p in palabras:
    if p and p.isdigit():
      c += 1
  return c

def contador_palabras_que_comienzan_con_vocal(texto):
  palabras = texto.lower().split(' ')
  c = 0
  for p in palabras:
    if p and p[0] in 'aeiouÃ¡Ã©Ã­Ã³Ãº':
      c += 1
  return c

def contador_palabras_que_comienzan_con_consonante(texto):
  palabras = texto.lower().split(' ')
  c = 0
  for p in palabras:
    if p and not p.isdigit() and p[0] not in ' aeiouÃ¡Ã©Ã­Ã³Ãº':
      c += 1
  return c

def palabras_en_infinitivo(texto):
  palabras = texto.lower().split(' ')
  s = ''
  for p in palabras:
    if p and p[-2:] in ['ar','er','ir']:
      s += p + ' '
  return s

print(f'Longitud del texto ðŸ‘‰ {len(texto)}')
print(f'Cantidad de palabras ðŸ‘‰ {contador_palabras(texto)} ')
print(f'Cantidad de oraciones ðŸ‘‰ {contador_oraciones(texto)}')
letra = input('Â¿Vocal o consonante?: ')
if letra.lower() == 'vocal':
  print(f'Cantidad de palabras que comienzan con una vocal ðŸ‘‰ {contador_palabras_que_comienzan_con_vocal(texto)}')
else:
  print(f'Cantidad de vocales {contador_palabras_que_comienzan_con_consonante(texto)}')
print(f'Cantidad de palabras que comienzan con mayÃºscula ðŸ‘‰ {contador_palabras_comienzan_mayuscula(texto)}')
print(f'Cantidad de palabras que son nÃºmeros ðŸ‘‰ {contador_de_carac_numericos(texto)}')
print(f'Palabras que terminan en infinitivo ðŸ‘‰ {palabras_en_infinitivo(texto)}')
```
