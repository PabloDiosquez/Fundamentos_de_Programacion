```python
class Equipo:
  def __init__(self, nombre, puntaje, goles):
    self.nombre = nombre
    self.puntaje = puntaje
    self.goles = goles
  
def cargar_equipos():
  equipos = []
  while True:
    nombre = input('Nombre: ')
    if nombre == 'fin': break
    puntaje = int(input('Puntaje: '))
    goles = int(input('Goles: '))

    equipos.append(Equipo(nombre,puntaje,goles))
  
  return equipos

def campeon(equipos):
  campeon = equipos[0]
  puntaje_max = equipos[0].goles

  for equipo in equipos:
    if equipo.puntaje > puntaje_max:
      campeon = equipo
      puntaje_max = equipo.puntaje
  
  return campeon

def main():
  equipos = cargar_equipos()
  print(f'Goles a favor del primer equipo de la lista: {equipos[0].goles}')
  print(f'Goles a favor del último equipo de la lista: {equipos[len(equipos) - 1].goles}')
  print(f'\nCampeón: {campeon(equipos).nombre}')
```

```python
class Alumno:
  
  def __init__(self, legajo, nombre, apellido, clave):
    self.legajo      = legajo
    self.nombre      = nombre
    self.apellido    = apellido
    self.clave = clave

  def __str__(self):
    return f'Legajo: {self.legajo} -- Nombre: {self.nombre} -- Apellido: {self.apellido} -- Clave: {self.clave}'

  def clave_valida(self):
    if len(self.clave) > 6 and self.clave[-1].isdigit():
      return True
    return False 

def procesar_alumnos():
  alumnos = list()
  while True:
    legajo = input('Legajo: ')
    if legajo == '0': break
    nombre = input('Nombre: ')
    apellido = input('Apellido: ')
    clave = input('Clave: ')

    alumnos.append(Alumno(legajo, nombre, apellido, clave))
  return alumnos
  
def imprimir_alumnos(alumnos: list):
  for alumno in alumnos:
    print(str(alumno))
  
def legajo_menor(alumnos: list) -> Alumno:
  alumnoMenor = alumno[0]
  legajo_menor = alumnos[0].legajo
  for alumno in alumnos:
    if len(alumno.legajo) < len(legajo_menor):
      alumnoMenor = alumno
      legajo_menor = alumno.legajo
  return alumnoMenor

def nombre_mas_largo(alumnos: list):
  alumno_mayor = alumnos[0]
  nombre_mas_largo = alumnos[0].nombre
  for alumno in alumnos:
    if len(alumno.nombre) > nombre_mas_largo:
      alumno_mayor = alumno
      nombre_mas_largo = alumno.nombre
  return alumno_mayor 

def verificar_clave(alumnos: list):
  for alumno in alumnos:
    if alumno.clave_valida():
      print(f'Legajo: {alumno.legajo} >>> Clave válida')
    else:
      print(f'Legajo: {alumno.legajo} >>> Error ❌ Clave inválida.')

def main():
  alumnos = procesar_alumnos()
  while True:
    print('''#### Menú de opciones ####
  1. imprimir los datos de todos los alumnos con el formato pedido en el punto a)
  2. imprimir los datos del alumno que tiene el legajo más chico.
  3. imprimir los datos del alumno que tiene el nombre más largo.
  4. Imprimir si las contraseñas de cada alumno cumplen con un tamaño mayor a 6 caracteres y terminan con un número.
  5. Salir
  ''')

    opcion = input('Ingrese opciòn: ')

    if opcion == '1':
      imprimir_alumnos(alumnos)
    elif opcion == '2':
      print(str(legajo_menor(alumnos)))
    elif opcion == '3':
      print(str(nombre_mas_largo(alumnos)))
    elif opcion == '4':
      verificar_clave(alumnos)
    elif opcion == '5':
      break
    else:
      print('Opción no válida. Intente de nuevo!')
  
  print('Programa terminado 💀')

main()
```

información de todos los cursos que se dictan en la misma. palabrasa cantidad de cursos

Importante: si la duración del curso es mayor a 360 minutos, NO se debe guardar la
información del mismo.
El programa debe permitir

a) Imprimir todos los cursos que tengan más de 5 clases

b) Imprimir los cursos que se dictan en un día de la semana indicado por el usuario.

c) Imprimir el curso más intensivo (definido por la cantidad de clases y la duración de
cada clase)

d) Imprimir los datos en que se dicta un curso indicado por el usuario.

e) Imprima todos los cursos que comienzan con una letra indicada por el usuario.

El programa escrito deberá mostrar un menú para que el usuario elija la opción deseada.
Importante: el sistema termina cuando el usuario elige la opción “salir”.

```python
class Curso:
  def __init__(self,nombre="",clases=0,dia="",hora=0,duracion=0):
    self.nombre = nombre
    self.clases = clases
    self.dia = dia
    self.hora = hora
    self.duracion = duracion
  
  def dia_de_semana(self):
    return self.dia

  def cantidad_clases(self):
    return self.clases

  def intensidad_del_curso(self):
    return self.clases*self.duracion

  def __str__(self):
    return f"""Nombre: {self.nombre} - Cantidad de clases: {self.clases} - Día: {self.dia} - Hora: {self.hora} - Duración: {self.duracion}
    """

  def primera_letra_nombre(self):
    return self.nombre[0]

def cargar_cursos():
  cursos = []
  c = int(input('Ingrese la cantidad de cursos: '))
  for i in range(c):
    print(f'Datos del curso {i+1}:')
    nombre = input('Nombre: ')
    clases = int(input('Cantidad de clases: '))
    dia = input('Día: ')
    hora = input('Hora: ')
    duracion = int(input('Duración: '))
    if duracion > 360: continue
 
    cursos.append(Curso(nombre,clases,dia,hora,duracion))
  return cursos

def datos_del_curso(lista,nombre):
  for e in lista:
    if e.nombre.lower() == nombre:
      return str(e)

def cursos_que_se_dictan_los(cursos,dia):
  for curso in cursos:
    if curso.dia.lower() == dia:
      print(curso.nombre)

def curso_mayor_intensidad(cursos):
  cursoMax = cursos[0].nombre
  mayorInt = cursos[0].intensidad_del_curso()
  for curso in cursos:
    if curso.intensidad_del_curso() > mayorInt:
      cursoMax = curso.nombre
      mayorInt = curso.intensidad_del_curso()
  return f'El curso de mayor intensidad es {cursoMax}'

def curso_empieza_con(cursos,letra):
  for curso in cursos:
    if curso.nombre[0].lower() == letra:
      print(curso.nombre)

def main():
  print('Cargar cursos:\n')
  cursos = cargar_cursos()
  print('Cursos con más de 5 clases:')
  for curso in cursos:
    if curso.clases > 5:
      print(curso.nombre)

  dia = input('Ingrese un día de la semana: ')
  cursos_que_se_dictan_los(cursos,dia.lower())

  print(curso_mayor_intensidad(cursos))
    
  nombre = input('Ingrese un nombre: ')
  print(datos_del_curso(cursos,nombre.lower()))

  letra = input('Ingrese una letra: ')
  curso_empieza_con(cursos,letra.lower())
    
main()
```

b) Realice una función que cree una lista con los datos del stock, donde de cada

c) Se requiere imprimir los datos de todos los artículos que necesitan refrigeración o
los artículos que no la necesitan, dependiendo del valor ingresado por el usuario
(R=requiere frío o N=no requiere frío).

d) Se necesita imprimir la cantidad de artículos que necesitan refrigeración y la
cantidad de artículos que no la necesitan.

e) Se requiere imprimir los datos de los artículos cuyo stock es menor a 5, de la siguiente
manera: Código, nombre del artículo, y nombre del proveedor.

f) Realice una función que utilizando la lista de productos creada en a) y en b) genere
una nueva lista con la información completa de los artículos que no requieren frío,
indicando además la cantidad existente.

Construir un menú, el menú deberá permitir ingresar 5 opciones, La opción 0 permite salir
del menú, el resto de las opciones permiten:
- imprimir los datos de los artículos que necesitan o no refrigeración, dependiendo
del valor ingresado por el usuario
- imprimir la cantidad de artículos que precisan refrigeración
- imprimir la cantidad de artículos que no requieren refrigeración
- imprimir los datos de los artículos cuyo stock es 0
- imprimir la información completa de los artículos que no requieren frío.

```python
# Clase Artículo
class Articulo:
  def __init__(self, codigo, nombre, descripcion, precio_venta, proveedor, refrigeracion):
    self.codigo        = codigo
    self.nombre        = nombre
    self.descripcion   = descripcion
    self.precio_venta  = precio_venta
    self.proveedor     = proveedor
    self.refrigeracion = refrigeracion

  def MostrarDatos(self):
    return f'Código: {self.codigo} -- Nombre: {self.nombre} -- Proveedor: {self.proveedor}'

  def requiere_refrigeracion(self):
    return self.refrigeracion == 'R'

# a)
def cargar_articulos():
  articulos = []
  while True:
    codigo = int(input('Código: '))
    nombre = input('Nombre: ')
    descripcion = input('Descripción: ')
    precio_venta = float('Precio de venta: ')
    proveedor = input('Proveedor: ')
    refrigeracion = input('¿Requiere refrigeración? R -- N')
    # Condición de corte
    if codigo == 999 and refrigeracion == 'N': break 
    
    # Agrego el artículo en la lista
    articulos.append(Articulo(codigo,nombre,descripcion,precio_venta,proveedor,refrigeracion))

  return articulos

# b) 
def stock(articulos):
  # Clave: artículo(objeto) 👉🏼 Valor: (código,cantidad)
  stock = {}

  # Recorro la lista de artículos, cargando el diccionario stock
  for articulo in articulos:
    if articulo not in stock:
      stock[articulo] = (articulo.codigo,1)
    else:
      stock[articulo][1] += 1
  
  return stock

def articulos_con_refrigeracion(articulos):
  # Lista con artículos que requieren refrigeración 
  articulos_con_refrigeracion = []
  for articulo in articulos: 
    if (articulo.requiere_refrigeracion()) and (articulo not in articulos_con_refrigeracion): # 👉 esto evita repeticiones de los artículos
      articulos_con_refrigeracion.append(articulo)
  # Recorro la lista sin repeticiones y muestro los datos de sus artículos
  for articulo in articulos_con_refrigeracion:
    print(articulo.MostrarDatos())

  return articulos_con_refrigeracion

def cantidad_articulos_con_refrigeracion(stock):
  # Dado que el stock guarda las cantidades, lo utilizo para calcular la cantidad de artículos con refrigeración
  total = 0
  for articulo in stock:
    if articulo.requiere_refrigeracion():
      total += stock[articulo][1]

  return total

def menu():
  #articulos = cargar_articulos()
  print("""
  ⚡imprimir los datos de los artículos que necesitan refrigeración 👉 Ingrese 1
  ⚡imprimir la cantidad de artículos que requieren refrigeración 👉 Ingrese 2
  ⚡imprimir la cantidad de artículos que no precisan refrigeración 👉 Ingrese 3
  ⚡imprimir los datos de los artículos cuyo stock es menor a 5 👉 Ingrese 4
  ⚡imprimir la información completa de los artículos que requieren frío. 👉 Ingrese 5
  🔚Ingrese 0 para salir del menú
  """)
  while True:
    opcion = input('Ingrese número de opción: ')
    if opcion.isdigit() and 0 <= int(opcion) <= 5: break
  
menu()
```

```python
texto = """Ni bien amainó
la tormenta, olvidó
las promesas hechas
otra vez!
Me matan, Limón!
Hijueputas, Limón!
Por los techos viene el bloque
otra vez!
Oh, otra vez!
Oh, otra vez!
Pasan las dos... Delatándolo,
apenas las dos... delatándolo a las dos
de la tarde y zumba feo algún moscón.
Voy sangrando, Limón! Padrecito de
Me la dieron, Limón! nuestros pobres que
sangran por las tejas como vos, Limón.
Oh, otra vez!
Oh, otra vez!
Nunca pude llorar,
ni soñar al dormir,
pero sabe que a las dos es tarde ya.
Todo el bloque, Limón, Nuestra gente, no,
baila y canta, mi Dios! no quiere escuchar
cumbias y merengues crueles otra vez!
Oh, otra vez!
Oh, otra vez!
Oh, otra vez!
Oh, otra vez!"""

def contador_palabras(texto):
  c = 0
  for carac in texto:
    if carac != ' ':
      c += 1
  return c

def contador_oraciones(texto):
  return len(texto.split('.\n'))

def contador_palabras_que_comienzan_con(texto,letra):
  palabras = texto.lower().split(' ')
  c = 0
  for p in palabras:
    if p and p[0] == letra: # when you split there will be an empty string at the end, and you can't access the first char char of this,
                            #  so just add a condition in the if:
                            # if word because 0-size word are False, and other True.
      c += 1
  return c 

def contador_de_palabra(texto,palabra):
  return texto.count(palabra)

def contador_palabras_comienzan_mayuscula(texto):
  palabras = texto.split(' ')
  c = 0
  for p in palabras:
    if p and p[0].isupper():
      c += 1
  return c

def contador_de_carac_numericos(texto):
  palabras = texto.split(' ')
  c = 0
  for p in palabras:
    if p and p.isdigit():
      c += 1
  return c

def contador_palabras_que_comienzan_con_vocal(texto):
  palabras = texto.lower().split(' ')
  c = 0
  for p in palabras:
    if p and p[0] in 'aeiouáéíóú':
      c += 1
  return c

def contador_palabras_que_comienzan_con_consonante(texto):
  palabras = texto.lower().split(' ')
  c = 0
  for p in palabras:
    if p and not p.isdigit() and p[0] not in ' aeiouáéíóú':
      c += 1
  return c

def palabras_en_infinitivo(texto):
  palabras = texto.lower().split(' ')
  s = ''
  for p in palabras:
    if p and p[-2:] in ['ar','er','ir']:
      s += p + ' '
  return s

print(f'Longitud del texto 👉 {len(texto)}')
print(f'Cantidad de palabras 👉 {contador_palabras(texto)} ')
print(f'Cantidad de oraciones 👉 {contador_oraciones(texto)}')
letra = input('¿Vocal o consonante?: ')
if letra.lower() == 'vocal':
  print(f'Cantidad de palabras que comienzan con una vocal 👉 {contador_palabras_que_comienzan_con_vocal(texto)}')
else:
  print(f'Cantidad de vocales {contador_palabras_que_comienzan_con_consonante(texto)}')
print(f'Cantidad de palabras que comienzan con mayúscula 👉 {contador_palabras_comienzan_mayuscula(texto)}')
print(f'Cantidad de palabras que son números 👉 {contador_de_carac_numericos(texto)}')
print(f'Palabras que terminan en infinitivo 👉 {palabras_en_infinitivo(texto)}')
```

```python
# Ejercicio 1a 💣

def son_iguales(texto1,texto2):
  if texto1[0] == texto2[0] and texto1[len(texto1)-1] == texto2[len(texto2)-1]:
    return 'IGUALES'
  return 'DISTINTOS'

def se_pueden_procesar(texto1,texto2):
  return len(texto1) > 2 and len(texto2) > 2

def main():
  texto1 = input('Texto 1: ').strip()
  texto2 = input('Texto 2: ').strip()
  if se_pueden_procesar(texto1,texto2):
    print(f'Resultado: {son_iguales(texto1,texto2)}')
  else:
    print('Los textos ingresados NO se pueden procesar')

main()
```

Programar una función de nombre suma_nros_pares() que calcule la suma de los números pares de una lista que recibe por parámetro.
Realice las correspondientes invocaciones con la lista [4,21,23,-2,8,0] y con []. ¿Qué devuelve en cada caso?

```python
def es_par(n):
  return not n % 2

def suma_nros_pares(numeros):
  suma = 0
  for numero in numeros:
    if es_par(numero):
      suma += numero
  return suma

print(suma_nros_pares([4,21,23,-2,8,0]))
print(suma_nros_pares([]))
```

Por último, deberá calcular el promedio de la cantidad de alumnos para los casos en los que haya más de 6 profesores.

```python
class Materia:
  def __init__(self,nombre,estudiantes,profesores):
    self.nombre      = nombre
    self.estudiantes = estudiantes
    self.profesores  = profesores

def cargar_materias():
  materias = []
  for x in range(3):
    nombre      = input('Nombre: ')
    estudiantes = input('Cantidad de estudiantes: ')
    profesores  = input('Cantidad de profesores: ')

    materias.append(Materia(nombre,estudiantes,profesores))

  return materias

def contar_profesores(materias,cantidad=5):
  total = 0
  for materia in materias:
    if materia.profesores >= cantidad:
      total += 1
  return total

def promedio_alumnos(materias):
  # Precondición: La lista de materias debe ser no vacía ⚠
  total = 0
  acum = 0
  for materia in materias:
    acum += materia.alumnos
    total += 1

  return round(acum/total,2)
```
