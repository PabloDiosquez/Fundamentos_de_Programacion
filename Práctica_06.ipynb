equipos que est√°n en la primera y √∫ltima posici√≥n de la lista.

c) Imprimir el nombre del equipo Campe√≥n de la lista generada en el punto a), el
equipo campe√≥n es aquel que sumo m√°s puntos.

```python
class Equipo:
  def __init__(self, nombre, puntaje, goles):
    self.nombre = nombre
    self.puntaje = puntaje
    self.goles = goles
  
def cargar_equipos():
  equipos = []
  while True:
    nombre = input('Nombre: ')
    if nombre == 'fin': break
    puntaje = int(input('Puntaje: '))
    goles = int(input('Goles: '))

    equipos.append(Equipo(nombre,puntaje,goles))
  
  return equipos

def campeon(equipos):
  campeon = equipos[0]
  puntaje_max = equipos[0].goles

  for equipo in equipos:
    if equipo.puntaje > puntaje_max:
      campeon = equipo
      puntaje_max = equipo.puntaje
  
  return campeon

def main():
  equipos = cargar_equipos()
  print(f'Goles a favor del primer equipo de la lista: {equipos[0].goles}')
  print(f'Goles a favor del √∫ltimo equipo de la lista: {equipos[len(equipos) - 1].goles}')
  print(f'\nCampe√≥n: {campeon(equipos).nombre}')
```

- Imprimir si las contrase√±as de cada alumno cumplen con un tama√±o mayor a 6
caracteres y terminan con un n√∫mero.

```python
class Alumno:
  
  def __init__(self, legajo, nombre, apellido, clave):
    self.legajo      = legajo
    self.nombre      = nombre
    self.apellido    = apellido
    self.clave = clave

  def __str__(self):
    return f'Legajo: {self.legajo} -- Nombre: {self.nombre} -- Apellido: {self.apellido} -- Clave: {self.clave}'

  def clave_valida(self):
    if len(self.clave) > 6 and self.clave[-1].isdigit():
      return True
    return False 

def procesar_alumnos():
  alumnos = list()
  while True:
    legajo = input('Legajo: ')
    if legajo == '0': break
    nombre = input('Nombre: ')
    apellido = input('Apellido: ')
    clave = input('Clave: ')

    alumnos.append(Alumno(legajo, nombre, apellido, clave))
  return alumnos
  
def imprimir_alumnos(alumnos: list):
  for alumno in alumnos:
    print(str(alumno))
  
def legajo_menor(alumnos: list) -> Alumno:
  alumnoMenor = alumno[0]
  legajo_menor = alumnos[0].legajo
  for alumno in alumnos:
    if len(alumno.legajo) < len(legajo_menor):
      alumnoMenor = alumno
      legajo_menor = alumno.legajo
  return alumnoMenor

def nombre_mas_largo(alumnos: list):
  alumno_mayor = alumnos[0]
  nombre_mas_largo = alumnos[0].nombre
  for alumno in alumnos:
    if len(alumno.nombre) > nombre_mas_largo:
      alumno_mayor = alumno
      nombre_mas_largo = alumno.nombre
  return alumno_mayor 

def verificar_clave(alumnos: list):
  for alumno in alumnos:
    if alumno.clave_valida():
      print(f'Legajo: {alumno.legajo} >>> Clave v√°lida')
    else:
      print(f'Legajo: {alumno.legajo} >>> Error ‚ùå Clave inv√°lida.')

def main():
  alumnos = procesar_alumnos()
  while True:
    print('''#### Men√∫ de opciones ####
  1. imprimir los datos de todos los alumnos con el formato pedido en el punto a)
  2. imprimir los datos del alumno que tiene el legajo m√°s chico.
  3. imprimir los datos del alumno que tiene el nombre m√°s largo.
  4. Imprimir si las contrase√±as de cada alumno cumplen con un tama√±o mayor a 6 caracteres y terminan con un n√∫mero.
  5. Salir
  ''')

    opcion = input('Ingrese opci√≤n: ')

    if opcion == '1':
      imprimir_alumnos(alumnos)
    elif opcion == '2':
      print(str(legajo_menor(alumnos)))
    elif opcion == '3':
      print(str(nombre_mas_largo(alumnos)))
    elif opcion == '4':
      verificar_clave(alumnos)
    elif opcion == '5':
      break
    else:
      print('Opci√≥n no v√°lida. Intente de nuevo!')
  
  print('Programa terminado üíÄ')

main()
```

Se desea realizar un programa para una universidad. El mismo deber√° permitir guardar la
informaci√≥n de todos los cursos que se dictan en la misma. palabrasa cantidad de cursos
existentes es ingresado por el usuario del sistema. De cada curso se debe conocer:
Nombre, Cantidad de Clases, D√≠a de Dictado, Hora de Comienzo, Duraci√≥n en minutos.
Importante: si la duraci√≥n del curso es mayor a 360 minutos, NO se debe guardar la
informaci√≥n del mismo.
El programa debe permitir

a) Imprimir todos los cursos que tengan m√°s de 5 clases

b) Imprimir los cursos que se dictan en un d√≠a de la semana indicado por el usuario.

c) Imprimir el curso m√°s intensivo (definido por la cantidad de clases y la duraci√≥n de
cada clase)

d) Imprimir los datos en que se dicta un curso indicado por el usuario.

e) Imprima todos los cursos que comienzan con una letra indicada por el usuario.

El programa escrito deber√° mostrar un men√∫ para que el usuario elija la opci√≥n deseada.
Importante: el sistema termina cuando el usuario elige la opci√≥n ‚Äúsalir‚Äù.

```python
class Curso:
  def __init__(self,nombre="",clases=0,dia="",hora=0,duracion=0):
    self.nombre = nombre
    self.clases = clases
    self.dia = dia
    self.hora = hora
    self.duracion = duracion
  
  def dia_de_semana(self):
    return self.dia

  def cantidad_clases(self):
    return self.clases

  def intensidad_del_curso(self):
    return self.clases*self.duracion

  def __str__(self):
    return f"""Nombre: {self.nombre} - Cantidad de clases: {self.clases} - D√≠a: {self.dia} - Hora: {self.hora} - Duraci√≥n: {self.duracion}
    """

  def primera_letra_nombre(self):
    return self.nombre[0]

def cargar_cursos():
  cursos = []
  c = int(input('Ingrese la cantidad de cursos: '))
  for i in range(c):
    print(f'Datos del curso {i+1}:')
    nombre = input('Nombre: ')
    clases = int(input('Cantidad de clases: '))
    dia = input('D√≠a: ')
    hora = input('Hora: ')
    duracion = int(input('Duraci√≥n: '))
    if duracion > 360: continue
 
    cursos.append(Curso(nombre,clases,dia,hora,duracion))
  return cursos

def datos_del_curso(lista,nombre):
  for e in lista:
    if e.nombre.lower() == nombre:
      return str(e)

def cursos_que_se_dictan_los(cursos,dia):
  for curso in cursos:
    if curso.dia.lower() == dia:
      print(curso.nombre)

def curso_mayor_intensidad(cursos):
  cursoMax = cursos[0].nombre
  mayorInt = cursos[0].intensidad_del_curso()
  for curso in cursos:
    if curso.intensidad_del_curso() > mayorInt:
      cursoMax = curso.nombre
      mayorInt = curso.intensidad_del_curso()
  return f'El curso de mayor intensidad es {cursoMax}'

def curso_empieza_con(cursos,letra):
  for curso in cursos:
    if curso.nombre[0].lower() == letra:
      print(curso.nombre)

def main():
  print('Cargar cursos:\n')
  cursos = cargar_cursos()
  print('Cursos con m√°s de 5 clases:')
  for curso in cursos:
    if curso.clases > 5:
      print(curso.nombre)

  dia = input('Ingrese un d√≠a de la semana: ')
  cursos_que_se_dictan_los(cursos,dia.lower())

  print(curso_mayor_intensidad(cursos))
    
  nombre = input('Ingrese un nombre: ')
  print(datos_del_curso(cursos,nombre.lower()))

  letra = input('Ingrese una letra: ')
  curso_empieza_con(cursos,letra.lower())
    
main()
```

Nos contratan para realizar un sistema para un kiosco. De cada art√≠culo se conoce
c√≥digo, nombre, una descripci√≥n, el precio de venta, el nombre del proveedor y si es
necesaria la refrigeraci√≥n del art√≠culo (pudiendo ser R=requiere fr√≠o o N=no requiere fr√≠o).

a) Realice una funci√≥n que cree una lista con informaci√≥n de art√≠culos de kiosco. La
carga de art√≠culos finaliza cuando se ingresa un art√≠culo cuyo c√≥digo = ‚Äú999‚Äù y que
a la vez no requiera fr√≠o.

b) Realice una funci√≥n que cree una lista con los datos del stock, donde de cada
articulo se conoce el c√≥digo y la cantidad existente.

c) Se requiere imprimir los datos de todos los art√≠culos que necesitan refrigeraci√≥n o
los art√≠culos que no la necesitan, dependiendo del valor ingresado por el usuario
(R=requiere fr√≠o o N=no requiere fr√≠o).

d) Se necesita imprimir la cantidad de art√≠culos que necesitan refrigeraci√≥n y la
cantidad de art√≠culos que no la necesitan.

e) Se requiere imprimir los datos de los art√≠culos cuyo stock es menor a 5, de la siguiente
manera: C√≥digo, nombre del art√≠culo, y nombre del proveedor.

f) Realice una funci√≥n que utilizando la lista de productos creada en a) y en b) genere
una nueva lista con la informaci√≥n completa de los art√≠culos que no requieren fr√≠o,
indicando adem√°s la cantidad existente.

Construir un men√∫, el men√∫ deber√° permitir ingresar 5 opciones, La opci√≥n 0 permite salir
del men√∫, el resto de las opciones permiten:
- imprimir los datos de los art√≠culos que necesitan o no refrigeraci√≥n, dependiendo
del valor ingresado por el usuario
- imprimir la cantidad de art√≠culos que precisan refrigeraci√≥n
- imprimir la cantidad de art√≠culos que no requieren refrigeraci√≥n
- imprimir los datos de los art√≠culos cuyo stock es 0
- imprimir la informaci√≥n completa de los art√≠culos que no requieren fr√≠o.

```python
# Clase Art√≠culo
class Articulo:
  def __init__(self, codigo, nombre, descripcion, precio_venta, proveedor, refrigeracion):
    self.codigo        = codigo
    self.nombre        = nombre
    self.descripcion   = descripcion
    self.precio_venta  = precio_venta
    self.proveedor     = proveedor
    self.refrigeracion = refrigeracion

  def MostrarDatos(self):
    return f'C√≥digo: {self.codigo} -- Nombre: {self.nombre} -- Proveedor: {self.proveedor}'

  def requiere_refrigeracion(self):
    return self.refrigeracion == 'R'

# a)
def cargar_articulos():
  articulos = []
  while True:
    codigo = int(input('C√≥digo: '))
    nombre = input('Nombre: ')
    descripcion = input('Descripci√≥n: ')
    precio_venta = float('Precio de venta: ')
    proveedor = input('Proveedor: ')
    refrigeracion = input('¬øRequiere refrigeraci√≥n? R -- N')
    # Condici√≥n de corte
    if codigo == 999 and refrigeracion == 'N': break 
    
    # Agrego el art√≠culo en la lista
    articulos.append(Articulo(codigo,nombre,descripcion,precio_venta,proveedor,refrigeracion))

  return articulos

# b) 
def stock(articulos):
  # Clave: art√≠culo(objeto) üëâüèº Valor: (c√≥digo,cantidad)
  stock = {}

  # Recorro la lista de art√≠culos, cargando el diccionario stock
  for articulo in articulos:
    if articulo not in stock:
      stock[articulo] = (articulo.codigo,1)
    else:
      stock[articulo][1] += 1
  
  return stock

def articulos_con_refrigeracion(articulos):
  # Lista con art√≠culos que requieren refrigeraci√≥n 
  articulos_con_refrigeracion = []
  for articulo in articulos: 
    if (articulo.requiere_refrigeracion()) and (articulo not in articulos_con_refrigeracion): # üëâ esto evita repeticiones de los art√≠culos
      articulos_con_refrigeracion.append(articulo)
  # Recorro la lista sin repeticiones y muestro los datos de sus art√≠culos
  for articulo in articulos_con_refrigeracion:
    print(articulo.MostrarDatos())

  return articulos_con_refrigeracion

def cantidad_articulos_con_refrigeracion(stock):
  # Dado que el stock guarda las cantidades, lo utilizo para calcular la cantidad de art√≠culos con refrigeraci√≥n
  total = 0
  for articulo in stock:
    if articulo.requiere_refrigeracion():
      total += stock[articulo][1]

  return total

def menu():
  #articulos = cargar_articulos()
  print("""
  ‚ö°imprimir los datos de los art√≠culos que necesitan refrigeraci√≥n üëâ Ingrese 1
  ‚ö°imprimir la cantidad de art√≠culos que requieren refrigeraci√≥n üëâ Ingrese 2
  ‚ö°imprimir la cantidad de art√≠culos que no precisan refrigeraci√≥n üëâ Ingrese 3
  ‚ö°imprimir los datos de los art√≠culos cuyo stock es menor a 5 üëâ Ingrese 4
  ‚ö°imprimir la informaci√≥n completa de los art√≠culos que requieren fr√≠o. üëâ Ingrese 5
  üîöIngrese 0 para salir del men√∫
  """)
  while True:
    opcion = input('Ingrese n√∫mero de opci√≥n: ')
    if opcion.isdigit() and 0 <= int(opcion) <= 5: break
  
menu()
```

Por ejemplo, sin ingresa 'HOLA' y 'HONDA' deber√° imprimir 'IGUALES'. Asegurarse de que cada palabra cuente con al menos 2 letras, sino imprimir 'Los textos ingresados NO se pueden procesar'

b) Pedir el ingreso de un texto y un n√∫mero por teclado, e imprimir el texto tantas veces como indica el n√∫mero. Aseg√∫rese que el texto sea no nulo, y el n√∫mero ingresado mayor a cero.

```python
# Ejercicio 1a üí£

def son_iguales(texto1,texto2):
  if texto1[0] == texto2[0] and texto1[len(texto1)-1] == texto2[len(texto2)-1]:
    return 'IGUALES'
  return 'DISTINTOS'

def se_pueden_procesar(texto1,texto2):
  return len(texto1) > 2 and len(texto2) > 2

def main():
  texto1 = input('Texto 1: ').strip()
  texto2 = input('Texto 2: ').strip()
  if se_pueden_procesar(texto1,texto2):
    print(f'Resultado: {son_iguales(texto1,texto2)}')
  else:
    print('Los textos ingresados NO se pueden procesar')

main()
```

Programar una funci√≥n de nombre suma_nros_pares() que calcule la suma de los n√∫meros pares de una lista que recibe por par√°metro.
Realice las correspondientes invocaciones con la lista [4,21,23,-2,8,0] y con []. ¬øQu√© devuelve en cada caso?

```python
def es_par(n):
  return not n % 2

def suma_nros_pares(numeros):
  suma = 0
  for numero in numeros:
    if es_par(numero):
      suma += numero
  return suma

print(suma_nros_pares([4,21,23,-2,8,0]))
print(suma_nros_pares([]))
```

Realice un programa a trav√©s del cual ingresar el nomobre de una persona, cantidad de estudiantes y cantidad de profesores. Se debr√°n cargar los datos de 10 materias. Luego, deber√° contar cu√°ntas materias tienen al menos 5 profesores.

Por √∫ltimo, deber√° calcular el promedio de la cantidad de alumnos para los casos en los que haya m√°s de 6 profesores.

```python
class Materia:
  def __init__(self,nombre,estudiantes,profesores):
    self.nombre      = nombre
    self.estudiantes = estudiantes
    self.profesores  = profesores

def cargar_materias():
  materias = []
  for x in range(3):
    nombre      = input('Nombre: ')
    estudiantes = input('Cantidad de estudiantes: ')
    profesores  = input('Cantidad de profesores: ')

    materias.append(Materia(nombre,estudiantes,profesores))

  return materias

def contar_profesores(materias,cantidad=5):
  total = 0
  for materia in materias:
    if materia.profesores >= cantidad:
      total += 1
  return total

def promedio_alumnos(materias):
  # Precondici√≥n: La lista de materias debe ser no vac√≠a ‚ö†
  total = 0
  acum = 0
  for materia in materias:
    acum += materia.alumnos
    total += 1

  return round(acum/total,2)
```
